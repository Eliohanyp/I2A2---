```python
############################################################################################################################################################

#IMPORTS E INSTALAÇÃO DAS BIBLIOTECAS UTILIZADAS
#!pip install tensorflow pandas seaborn matplotlib wordcloud openpyxl

from tensorflow.keras.preprocessing.text import Tokenizer

import openpyxl

from wordcloud import WordCloud

import pandas as pd

import seaborn as sn

import matplotlib.pyplot as plt
from matplotlib import *
from matplotlib.colors import ListedColormap

############################################################################################################################################################

#LENDO O ARQUIVO DE TEXTO I (.txt)
arquivo_texto1 = 'Base Text - Matrix Representation I - v2.txt'
with open (arquivo_texto1, encoding='utf-8') as texto1:
    conteudo_texto = texto1.readlines()

#LIMPEZA DO TEXTO I      
texto_limpo = "\n',".join(conteudo_texto).replace("\n',","").replace("',", "\n").replace("'\n'", "\n").replace(" '","").replace(" ,","").replace("\n"," ")
#DEIXANDO TUDO MINUSCULO      
texto_matriz = texto_limpo.lower()
#SEPARANDO POR FRASES
texto_matriz = texto_limpo.split('.')

#CRIEI O TOKENIZER
tokenizer = Tokenizer()

# UTILIZA O 'texto_matriz' PARA CRIAR O DICIONARIO DE TOKENS
tokens = tokenizer.fit_on_texts(texto_matriz)
token_texto = tokenizer.word_index

#TRANFORMANDO O "text_matriz" EM MATRIZ 
matriz = tokenizer.texts_to_sequences(texto_matriz)

############################################################################################################################################################

#LENDO O ARQUIVO DE TEXTO II (.txt)
arquivo_texto2 = 'Base Text - Matrix Representation II - v2.txt'
with open (arquivo_texto2, encoding='utf-8') as texto2:
    conteudo_texto2 = texto2.readlines()

#LIMPEZA DO TEXTO II    
texto_limpo2 = "\n',".join(conteudo_texto2).replace("\n',","").replace("',", "\n").replace("'\n'", "\n").replace(" '","").replace(" ,","").replace("\n"," ")
#DEIXANDO TUDO MINUSCULO 
texto_matriz2 = texto_limpo2.lower()
#SEPARANDO POR FRASES
texto_matriz_paragrafo = texto_limpo2.split('.')

#CRIEI O TOKENIZER2
tokenizer2 = Tokenizer()

# UTILIZA O 'texto_matriz2' PARA CRIAR O DICIONARIO DE TOKENS
tokens2 = tokenizer2.fit_on_texts(texto_matriz_paragrafo)
token_texto2 = tokenizer2.word_index
#TRANFORMANDO O "text_matriz2" EM MATRIZ 
matriz2 = tokenizer2.texts_to_sequences(texto_matriz_paragrafo)

############################################################################################################################################################
```

```python
#SALVA A "matriz" EM UM DATAFRAME E PREENCHE AS COLUNAS VAZIAS COM "0.0"
df = pd.DataFrame(matriz).fillna(0)

#SALVA O DATAFRAME EM UM ARQUIVO .XLSX COM O NOME "Matrix Representation I.xlsx"
df.to_excel("Matrix Representation I.xlsx") 
```

```python
#SALVA A "matriz" EM UM DATAFRAME E PREENCHE AS COLUNAS VAZIAS COM "0.0"
df2 = pd.DataFrame(matriz2).fillna(0)

#SALVA O DATAFRAME EM UM ARQUIVO .XLSX COM O NOME "Matrix Representation II.xlsx"
df2.to_excel("Matrix Representation II.xlsx") 
```

```python
#TEXTO 1 - Base Text - Matrix Representation I - v2.txt

#CRIA UMA LISTA DE STOPWORDS EM LATIN E RETIRA AS PALAVRAS E SALVA ISSO NO NOME "texto_sem_stopword2"
stopword_latin = ['a','ab','ac', 'ad', 'at', 'atque', 'aut', 'autem', 'cum', 'de', 'dum', 'e',
                       'erant', 'erat', 'est', 'et', 'etiam', 'ex', 'haec', 'hic', 'hoc', 'in', 'ita', 'me', 'nec', 'neque', 
                       'non', 'per', 'qua', 'quae', 'quam', 'qui', 'quibus', 'quidem', 'quo', 'quod', 're', 'rebus', 'rem', 'res',
                       'sed', 'si', 'sic', 'sunt', 'tamen', 'tandem', 'te', 'ut', 'vel']

texto_sem_stopword = [k for k in token_texto if not k in stopword_latin] 

texto_sem_stopword2 = " ".join(texto_sem_stopword)

#CRIA UMA NUVEM DE PALAVRAS COM AS PALAVRAS UTILIZADAS NO TEXTO
mapa_cor = ListedColormap(['orange', 'green', 'red', 'blue'])

nuvem = WordCloud(background_color = 'white', max_words = 1000, colormap=mapa_cor)

nuvem = nuvem.generate(texto_sem_stopword2)
plt.figure(figsize=(15,15))
plt.imshow(nuvem)
plt.axis('off')
plt.show()
```

```python
#TEXTO 2 - Base Text - Matrix Representation II - v2.txt

#CRIA UMA LISTA DE STOPWORDS EM LATIN E RETIRA AS PALAVRAS E SALVA ISSO NO NOME "texto_sem_stopword2_II"
stopword_latin_II = ['a','ab','ac', 'ad', 'at', 'atque', 'aut', 'autem', 'cum', 'de', 'dum', 'e',
                       'erant', 'erat', 'est', 'et', 'etiam', 'ex', 'haec', 'hic', 'hoc', 'in', 'ita', 'me', 'nec', 'neque', 
                       'non', 'per', 'qua', 'quae', 'quam', 'qui', 'quibus', 'quidem', 'quo', 'quod', 're', 'rebus', 'rem', 'res',
                       'sed', 'si', 'sic', 'sunt', 'tamen', 'tandem', 'te', 'ut', 'vel']

texto_sem_stopword_II = [k for k in token_texto2 if not k in stopword_latin_II] 

texto_sem_stopword2_II = " ".join(texto_sem_stopword_II)

#CRIA UMA NUVEM DE PALAVRAS COM AS PALAVRAS UTILIZADAS NO TEXTO 2 
mapa_cor_II = ListedColormap(['orange', 'green', 'red', 'blue'])

nuvem_II = WordCloud(background_color = 'white', max_words = 1000, colormap=mapa_cor_II)

nuvem_II = nuvem.generate(texto_sem_stopword2_II)
plt.figure(figsize=(15,15))
plt.imshow(nuvem_II)
plt.axis('off')
plt.show()
```
